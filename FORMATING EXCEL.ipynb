```python
import pandas as pd
import time
from datetime import datetime
sheet = ['WeConnect','RSA','HR Data','HGA per JC','Joyride']
```

```python
with pd.ExcelWriter('SRV SCORE ' + time.strftime("%Y%m%d-%H%M%S") + '.xlsx') as writer:
    for i in sheet:
        print(i)
        df = pd.read_excel('Copy of MyScore Power BI UPLOAD FORMAT.xlsx', sheet_name=i)
        df['MONTH_NAME'] = now.strftime("%Y %B")[0:8]
        df = df.replace('#DIV/0!', 0) 
        df = df.replace('#DIV/0! ', 0)
        df = df.replace('#REF!', 0) 
        df = df.fillna(0)
        if(i == 'WeConnect'):
            df.drop(df.columns[[0,13,14]],axis = 1, inplace=True)
            df = df.rename(columns=df.iloc[0])
            print(df.shape)
        elif(i == 'RSA'):
            df = df.rename(columns=df.iloc[0])
            df.drop(df.columns[[0]],axis = 1, inplace=True)
            df.drop([0], axis=0, inplace=True)
            print(df.shape)
        elif(i == 'HR Data'):
            df = df.rename(columns=df.iloc[1])
            df.drop(df.columns[[0]],axis = 1, inplace=True)
            df.drop([0,1], axis=0, inplace=True)
            print(df.shape)
        elif(i == 'HGA per JC'):
            df = df.rename(columns=df.iloc[1])
            df.drop(df.columns[[0]],axis = 1, inplace=True)
            df.drop([0,1], axis=0, inplace=True)
            print(df.shape)
        elif(i == 'Joyride'): 
            df = df.rename(columns=df.iloc[0])
            df.drop([0], axis=0, inplace=True)
            df.drop(df.columns[[0]],axis = 1, inplace=True)
            df.rename(columns = {'Reporting Code':'DEALER_CODE'}, inplace = True)
            df.rename(columns = {'Joyride Sale Monthly Actual':'SALES'}, inplace = True)
            df.rename(columns = {'Description':'Year'}, inplace = True)
            df.rename(columns = {'Zone':'Month'}, inplace = True)
            df.rename(columns = {'Area Office':'Package_Type'}, inplace = True)
            df.rename(columns = {'Joyride Sale Monthly Target':'Enterdate'}, inplace = True)
            df.rename(columns = {'Joyride Achievement in %':'EntryBy'}, inplace = True)
            df.rename(columns = {'Type of Network':'MONTH_NAME'}, inplace = True)
            print(df.shape)
        df.to_excel(writer, sheet_name=i, index=False)
```
